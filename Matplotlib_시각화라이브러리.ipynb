### 교통사고가 가장 많이 발생하는 시간대를 알아보고 시각화 해보자.

```python
# '0~2','3~5','6~8','9~10','11~13', .. '21~23' 
```

```python
rng=data['발생년월일시']
bins=[-1,3,6,9,12,15,18,21,24]
labels=['0-2','3-5','6-8','9-11','12-14','15-17','18-20','21-23']
```

```python
#data.info()
data['발생년월일시']%100    ### 십의 자리까지 구하기 위해서 100으로 나눈 나머지를 구한다
```

```python
bins=[-1,2,5,8,11,14,17,20,23]  ### 실질적으로 데이터를 가르는 구간
labels=['0-2','3-5','6-8','9-11','12-14','15-17','18-20','21-23']### 라벨은 그냥 우리가 지정해주는 구간
result=pd.cut(data['발생년월일시']%100,bins=bins,labels=labels)
```

```python
rs=result.value_counts().sort_index()
rs
```

```python
plt.figure(figsize=(5,5) )
plt.pie(rs, labels=labels,autopct='%.2f')  ### 파이차트 원 차트 그려본다
plt.show()
```
