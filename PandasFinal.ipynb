### 범죄현황 데이터 실습

```python
### 불러오기 crime  
df2015=pd.read_csv('data/crime/2015.csv', encoding='euc-kr',index_col='관서명')
df2016=pd.read_csv('data/crime/2016.csv', encoding='euc-kr',index_col='관서명')
df2017=pd.read_csv('data/crime/2017.csv', encoding='euc-kr',index_col='관서명')
```



살인,강도,강간,절도,폭력 컬럼의 데이터 수를 합하여 총계라는 컬럼을 추가해보자

```python
total_2015=df2015['살인']+df2015['강도']+df2015['강간·강제추행']+df2015['절도']+df2015['폭력']
#total_2015.shape
df2015['총계']=total_2015
df2015
```

```python
total_2016=df2016['살인']+df2016['강도']+df2016['강간·강제추행']+df2016['절도']+df2016['폭력']

df2016['총계']=total_2016
df2016
```

```python
total_2017=df2017['살인']+df2017['강도']+df2017['강간·강제추행']+df2017['절도']+df2017['폭력']

df2017['총계']=total_2017
df2017
```

```python
### 풀이
df2015['총계']=df2015.sum(axis=1)
df2016['총계']=df2016.sum(axis=1)
df2017['총계']=df2017.sum(axis=1)
```

아래와 같은 형태로 만들기


```python
res15=df2015 [ df2015['구분']=='발생건수' ]
res16=df2016 [ df2016['구분']=='발생건수' ]
res17=df2017 [ df2017['구분']=='발생건수' ]

```

```python
res17_1=res17.drop( ['광주지방경찰청'] )
res17_1.info
```

```python
num1=(res16['총계'] -  res15['총계']) /res16['총계']*100
#num1
num2=(res17_1['총계'] -  res16['총계']) /res17_1['총계']*100

res=pd.concat( [ res15['총계'],  num1, res16['총계'], num2, res17_1['총계'] ] ,axis=1)
#res.info()
res.columns='2015총계','2015-2016 증감율','2016총계','2016-2017 증감률','2017총계'
res
```

```python
### 풀이
s1=df2015 [ df2015['구분']=='발생건수'].loc[:,'총계']
s1.name='2015총계'
#s1
```

```python
s2=df2016 [ df2016['구분']=='발생건수'].loc[:,'총계']
s2.name='2016총계'
#s2
```

```python
s3=df2017 [ df2017['구분']=='발생건수'].loc[:,'총계']
s3.name='2017총계'
#s3
```

```python
s4=(s2-s1)/s1*100
s4.name='2015-2016 증감율'
s4
```

```python
s5=(s3-s2)/s2*100
s5.name='2016-2017 증감율'
s5
```

```python
final=pd.concat( [s1,s4,s2,s5,s3 ] ,axis=1)
final.iloc[:-1]  ### 자료가 없는 광주지방경찰청 을 빼고 슬라이싱
```

```python
final.drop('광주지방경찰청',inplace=True)
```

#### groupby

```python
### 딕셔너리를 이용한 데이터프레임 생성하기    키 값들이 컬럼이 된다 
df = pd.DataFrame({
    'key1' : ['A', 'A', 'B', 'B', 'A'],
    'key2' : ['one', 'two', 'one', 'two', 'one'],
    'data1' : [1, 2, 3, 4, 5],
    'data2' : [10, 20, 30, 40, 50]
})
```

```python
df
```

```python
y=df[[ 'key1','data2']].groupby('key1')
```
