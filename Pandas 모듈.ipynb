```python
#csv파일 불러오기
population_number=pd.read_csv('data/population_number.csv',index_col='도시', encoding='euc-kr')
population_number
```

```python
#데이터의 개수 세기
population_number.count()
```

```python
# 정렬       ### 기본값은 오름차순이다.   점점 커지는 순
population_number['2010'].sort_values()
```

```python
### 내림차순            
###어센딩이 True면 오름차순이 된다 사실 오름차순은 기본값이다  결측치는 차순에 관계없이 가장아래로 간다
population_number['2010'].sort_values(ascending=False)
```

```python
### 2010년 기준으로 정렬된다  내림차순
population_number.sort_values(by='2010', ascending=False)
```

```python
#지역, 2015년 인구수 기준으로 정렬   ### 기준이 2개 이상이 될 때. 기준을 리스트형태로 넣는다
population_number.sort_values(by=['지역','2015'])
```

```python
### 결측치 채우기    ~.fillna(밸류=~)
population_number['2005'].fillna(value=0)
```

```python
### 인플레이스 속성.   ~.fillna(value=값, inplace=True) 
### 기본값은 False이다   True사용하면 원본데이터에 바로 적용된다. 
population_number.fillna(value=0, inplace=True)
population_number
```

```python
# 중복 데이터 수 세기  ~.value_counts()
s1=pd.Series([1,1,5,5,3,4,2,2,2,3,4,4,4,5,6,7,7,7,1,2,2,2,3])
s1.value_counts()
```

```python
### 중복카운트 첫번째 방법
population_number['2005'].value_counts()
```

```python
### 중복카운트 2번째 방법   서브셋
population_number.value_counts(subset=['2005'])
```

```python
# 카테고리만들기   pd.cat( ) 카테고리를 만들어낸다
ages=[0,2,10,21,23,37,31,61,20,41,32,100]
bins=[-1,15,25,35,60,100]
labels=['미성년자','청년','중년','장년','노년']

cats=pd.cut(ages,bins,labels=labels)
```

```python
cats
```

```python
type(cats)
```

```python
### 카테고리 인덱스 확인  ~.categories
cats.categories
```

```python
ageArr=pd.DataFrame(ages,columns=['ages'])
ageArr['cats']=cats
ageArr
```

```python
#score.csv파일 불러오기
#한글이 포함되어 있기때문에 인코딩해주기 
#인덱스컬럼은 과목컬럼으로 설정하기
score=pd.read_csv('data/score.csv',index_col='과목', encoding='euc-kr')
score
```

```python
# 합계 구하기     기본 ~>열별(컬럼별)로 합계를 구하는 것과 같다
score.sum()
```

```python
score.sum().sort_values(ascending=False)
```

```python
### 액시스 매개변수를 사용하면 연산기준(행/열)을 바꿔줄 수 있다.   none들은 0이라고 생각하면 된다
score.sum(axis=1)
```

```python
sum_s=score.sum(axis=1)
score['합계']=sum_s
score
```

```python
### 위와 같은 방법은 score 합계가 들어가면서 연산을 할때마다 합계값이 누적되는 문제가 생긴다
score['합계']=score.loc[:,:'4반'].sum(axis=1)
score
```

```python
# 평균 구하기 
score.mean(axis=1)
```

과목별 평균


```python
#과목별 평균 구해서 추가하기   ### loc 인덱서를 사용해야 올바른 값이 나온다
score['평균']=score.loc[:,:'4반'].mean(axis=1)
score
```

```python
# 행 추가하기 (loc인덱서 활용)
score.loc['반평균']=score.loc[:'과학',:'4반'].mean()
score
```
