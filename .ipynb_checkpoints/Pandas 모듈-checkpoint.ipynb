```python
import pandas as pd
```

### Series 사용

```python
#series 생성
obj1 =pd.Series([4,7,-5,3])
obj1
```

```python
# 인덱스를 지정하여 생성
obj2 = pd.Series( [4,7,-5,3], index=['d','b','a','c'])
obj2
```

```python
### 이미 만들어진 리스트에도 인덱스를 부여할 수 있다
obj1.index=['a','b','c','d']
obj1
```

```python
# 딕셔너리를 이용해서 Series 생성       ### 키값이 인덱스가 된다
dic={'a':10,'b':15,'c':13,'d':5}
obj3=pd.Series(dic)
obj3
```

```python
# Series 값 확인    ### 시리즈클래스의 밸류값을 확인하면 넘파이배열의 형태로 결과를 확인할 수 있다
obj2.values
```

```python
#Series인덱스 확인
obj2.index
```

```python
#Series타입 확인
obj2.dtype
```

연습문제


```python
#[9904312, 3448737, 2890451,2466052]
s2015=pd.Series( [9904312, 3448737, 2890451,2466052],index=['서울','부산','인천','대구']  )
s2015
```

```python
# Series에 이름지정하기
s2015.name='인구'
s2015
```

```python
# 인덱스 이름 지정
s2015.index.name='도시'
s2015
```

```python
#Series 연산
s2015/1000000
```

```python
# 동일한 라벨을 가지고 있는 데이터끼리 연산
s2015+s2015   ### 위치라기 보다는 동일한 라벨의 데이터들을 연산하는 것이라고 한다
```

```python
s2015>=3000000  ### 조건식연산??? 이 가능하기 때문에 불리언으로 반환이 된다
```

```python
# Series 인덱싱  ### 2가지 방법이 가능하다
s2015[0] , s2015['서울']
```

```python
s2015[3], s2015['대구']
```

```python
### 시리즈형태를 유지하면서 보고싶다면 대괄호 2번이면 된다   
s2015[['부산']]
```

```python
s2015[[0,2,1,3]]   ### 대괄호가 1개있을때는 일반적인 int타입이기때문에 위쪽과 같은 연산은 불가능
```

```python
# Series 슬라이싱   -> 결과값은 무조건 Series 클래스로 출력 
###부산에서 인천까지 
s2015[1:3]
```

```python
s2015['부산':'인천']  ### 숫자형인덱스들의 마지막은 +1 이었지만, 라벨을 이용할때는 원하는 곳까지만 넣어주면 된다
### '부산' : '대구' 가 아니다
```

```python
s2015[0]
```

```python
s2015[:1]
```

```python
# Boolean색인   ### True에 해당하는 값들만 출력된다.  비교연산을 넣어주게되면
s2015[s2015>= 2500000 ]
```

```python
s2015[s2015<= 5000000 ]
```

```python
s2015[ (s2015>= 2500000) & ( s2015<= 5000000)]
```

```python
#2010년 인구수
dic2={ '서울':9631482,'부산':3393191,'인천':2632035,'대전':1490158}
s2010=pd.Series(dic2)
s2010.name='2010년 인구'
s2010.index.name='도시'
s2010
```

```python
# 2010년에서 2015년에 증가한 인구수   ### NaN을 결측치 를 의미한다
ds=s2015-s2010
ds
```

```python
#null값이 아닌 데이터만 확인
ds.notnull()   ### 값이 없는것이 False 있는 것이 True로 빠져나온다
ds[ds.notnull()]
```

```python
#null값인 데이터만 확인
ds[ds.isnull()]
```

```python
#2015년도와 2010년도의 인구증가율(%) 계산
rs= (s2015-s2010)/s2010*100
rs[rs.notnull()]
```

```python
#Series 데이터 갱신, 추가, 삭제
### 갱신
rs['부산']=1.6
### 추가
rs['광주']=1.5
###삭제  예약어 del을 사용한다
del rs['대구']
rs
```

### DataFrame 사용

```python
#딕셔너리를 이용해서 생성  2차원 과 리스트형태
dic3={ '2015':[9904312, 3448737, 2890451, 2466052],'2010':[9631482, 3393191, 2632035, 2431774]}
###데이터프레임1
df1=pd.DataFrame(dic3)
df1.index=['서울','부산','인천','대구']
df1
```

```python
#리스트 이용해서 생성       ### 2행 4열의 구조
li1=[ [9904312, 3448737, 2890451, 2466052], [9631482, 3393191, 2632035, 2431774]]
df2=pd.DataFrame(li1)
df2.index=['2015','2010']
df2.columns=['서울','부산','인천','대구']
df2
```

```python
# 행과 열 구조 변경하기
df2=df2.T  
```

```python
#DataFrame 값 확인   넘파이배열 형태로 
df1.values
```

```python
#DataFrame 인덱스 확인  
df1.index
```

```python
#컬럼 확인
df1.columns
```

```python
# 인덱싱 -> 열 인덱싱 ~~> 결과값은 Series클래스로 출력
#결과값인 Series클래스의 이름은 컬럼의 이름과 같다
df1['2015']
```
