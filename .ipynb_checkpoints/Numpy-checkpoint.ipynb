```python
#boolean 색인       불리언색인을 할때는 요소의 갯수가 같아야 한다  # 각 요소에 참 거짓을 판별해서 들어간 배열이 만들어진다
score=np.array([80,75,66,98,71,83,45,89,99,90])
arr_bool= score>=60
arr_bool
#스코어 에서 데이터를 가져오겠다 는 뜻
score[arr_bool]
```

```python
score=np.array([80,75,66,98,71,83,45,89,99,90])
arr_bool= score>=60
arr_bool
#스코어 에서 데이터를 가져오겠다 는 뜻
score[~arr_bool]    # 물결표를 쓰면 연산을 뒤집을 수 있다
```

```python
score[score>=70]
```

```python
score=np.array([80,75,66])
arr_bool=np.array([True,False,True])
score[arr_bool]
```

```python
data1 = np.loadtxt('data/ratings.dat',delimiter='::',dtype=np.int64)
data1
```

```python
# 데이터 정보 확인하기
print(data1.ndim)  # 데이터의 차원 디멘션 확인
print(data1.shape)  # 데이터의 shape
print(data1.size)  # 데이터의  행과 열을 곱한 숫자.   데이터의 총 숫자
```

```python
# 전체 평점 평균 구하기   평점이 2번 인덱스이다
data1[:,2].mean ()
```

```python
#사용자별 평점 평균 구하기
### 사용자별로 구해보기   중간에 빠진 사용자들이 있을 수 있다는 점, 하나하나 눈으로 하기 불편하다는 점
### 유니크 unique   중복된 것 없애고 1가지씩만 취하기
unique_id=np.unique(data1[:,0])  #유니크한 사용자 아이디 추출 6040개
### 불리언 조건을 바꿔주면 되는데, 이것은 Unique_id에 들어있다

userId_ratings=[]
for user_id in unique_id:
    #print( user_id,data1[data1[:,0]==user_id,2].mean())   ### 출력하고 끝
    userId_ratings.append([user_id , data1[data1[:,0]==user_id,2].mean()]) ### 배열에 담는다 
    
```

```python
#영화별 평점 평균 구하기
```

```python
#data1[:,1]==1  ## 불리언 배열로 생성

itemId_ratings=[]
unique_item_id= np.unique( data1[:,1])
for item_id in unique_item_id:
    #print( data1[ data1[:,1]==item_id,2 ].mean() )
    ratings_mean= data1[ data1[:,1]==item_id,2 ].mean() 
    itemId_ratings.append([item_id,ratings_mean] )

#np.unique( data1[:,1]).shape

#data1[ data1[:,1]==3,2 ].mean()
```

```python
arr_item_ratings=np.array(itemId_ratings)
arr_item_ratings    ### 이대로는 넘파이에서 과학적표기법, 지수표현식 을 쓰기때문에 출력이 이상하게 나온다
np.set_printoptions(precision=2,suppress=True)  ###자릿수를 의미한다  서프레스 억제한다
#itemId_ratings
arr_item_ratings 
```
