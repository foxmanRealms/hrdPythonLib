### Numpy Basic

```python
#array 생성하기 : 1차원
li1=[2,7,8,1,0,5]  #리스트타입
arr1=np.array(li1)
arr1
```

```python
arr2=np.array([1,2,3,4,5] )
arr2
```

```python
# arrr생성하기 : 2차원
arr3=np.array(  [[1,2,3],[4,5,6]  ])
arr3
```

```python
#배열 크기 확인하기 (행, 열)
print(arr2.shape)
print(arr3.shape)
```

```python
# 배열의 전체 요소 갯수 확인하기
print(arr2.size)
print(arr3.size)
```

```python
#배열의 차원 확인하기  디멘션 확인
print(arr2.ndim)
print(arr3.ndim)
```

```python
#0으로 초기화 해서 배열 생성하기
arr_zeros=np.zeros((3,4))
arr_zeros
```

```python
# 1로 초기화해서 배열 생성하기
arr_ones=np.ones((3,4))
arr_ones
```

```python
# 특정한 값으로 초기화해서 배열 생성하기 .   0 과 1 빼고는 다른함수를 쓴다
arr_full=np.full((5,5),7)
arr_full
```

```python
#1~50이 담긴 배열 생성하기
arr4=np.arange(1,51,1)   # 1부터 시작해서  51보다 작을때까지 1씩 증가하는형태로 만들꺼야~
arr4
```

```python
# 랜덤 값 배열 생성하기
arr5=np.random.rand(2,3)
arr5
```

```python
# 랜덤한 정수값으로 배열 생성하기
arr6=np.random.randint(2,10,size=(3,4) )
arr6
```

```python
#타입을 지정해서 배열 생성하기
li2=[1.2,2.5,7.9]
arr_type=np.array(li2,dtype=np.int64)
arr_type
```

```python
#타입변경하기
arr_astype=arr_type.astype('float64')
arr_astype
```

```python
arr7=np.arange(1,51,1).reshape(5,10)   # 리쉐잎을 할때에는 서로 곱셈했을때 사이즈가 같아야한다. 안그러면 오류난다
arr7
```

### 인덱싱,슬라이싱

```python
# array 인덱싱
arr8=np.array([2,7,8,9,1,4])
print(arr8[3])

arr9=np.array( [ [1,2,4],[5,3,8]])
print(arr9[0])
print(arr9[0][2])
print(arr9[0,2])
print(arr9[1,2]) #8을 원할때     , 컴마를 이용해서 불러올 수 있다
```

```python
# array 슬라이싱
arr10=np.arange(10)
arr10[2:7]    #2부터 6까지 가져오고 싶을때
```

```python
arr11=np.arange(1,51,1).reshape(5,10)
arr11[0:2]   # 0 1 번째 행만 가져오고 싶을 때
arr11[0:2,2]# 0 1번째 행에서 3과 13만 가져오고 싶을때
```

arr11


```python
arr11[1:4,2:7]  # 컴마 앞은 행  뒤는 열이다
```

### Array Operation

```python
#boolean 색인       불리언색인을 할때는 요소의 갯수가 같아야 한다  # 각 요소에 참 거짓을 판별해서 들어간 배열이 만들어진다
score=np.array([80,75,66,98,71,83,45,89,99,90])
arr_bool= score>=60
arr_bool
#스코어 에서 데이터를 가져오겠다 는 뜻
score[arr_bool]
```

```python
score=np.array([80,75,66,98,71,83,45,89,99,90])
arr_bool= score>=60
arr_bool
#스코어 에서 데이터를 가져오겠다 는 뜻
score[~arr_bool]    # 물결표를 쓰면 연산을 뒤집을 수 있다
```

```python
score[score>=70]
```

```python
score=np.array([80,75,66])
arr_bool=np.array([True,False,True])
score[arr_bool]
```
