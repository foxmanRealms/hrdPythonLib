#### 나이별 카테고리 만들기

```python
### 마지막에 연령대 카테고리를 만들어버리는 것이다
ages=train['Age']
bins=[-1,15,25,35,60,100]
labels=['미성년자','청년','중년','장년','노년']

age_cats=pd.cut(ages,bins,labels=labels)
age_cats
```

```python

train['Age_cats']=age_cats
train
```

```python
train['Age_cats'].value_counts()
```

#### 6. 생존자 중 카테고리별 비율(미성년자,청년,중년,장년,노년 승객 비율)
